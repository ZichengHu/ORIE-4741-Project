```python
import numpy as np
import pandas as pd
# from wordcloud import WordCloud, STOPWORDS 
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot') # optional: for ggplot-like style

import datetime 
import seaborn as sns
```

```python
df = pd.read_csv('AB_NYC_2019.csv')

df = df.drop(columns = ['id', 'host_name','name'])
#df = df.fillna({'reviews_per_month':0}, inplace=True)
```

```python
#check the relationship between neighborhoods and listed prices
num_properties_nh = df['neighbourhood_group'].value_counts()
avg_price_nh = (df.groupby(['neighbourhood_group'],as_index = False).mean())[['neighbourhood_group','price']]
avg_price_nh = avg_price_nh.set_index(['neighbourhood_group'])
ax = avg_price_nh.plot(kind = 'bar', figsize = (8,5))
ax.set_xticklabels(avg_price_nh.index, rotation = 45)
ax.set_title('Average Price Per Night ')
ax.set_ylabel('$')
ax.set_xlabel('Neighbourhood Group')
```

```python
ax = sns.violinplot(data=df[df['price'] < 1000], x='neighbourhood_group', y='price',figsize = (8,5))
```

```python
df = df.drop(columns = ['host_id'])
df['neighbourhoods'] = df['neighbourhood_group'] + " " + df['neighbourhood']
df = df.drop(columns = ['neighbourhood_group', 'neighbourhood'])
d = []
current_date = datetime.datetime.strptime('2019-10-31', '%Y-%m-%d')

for i in range(len(df)):
    if pd.isna(df['last_review'][i]):
        d.append(np.NaN)
    else:
        last_review = datetime.datetime.strptime(df['last_review'][i], '%Y-%m-%d')
        d.append((current_date - last_review).days)

df['days_since_last_review'] = d
#df['days_since_last_review'] = df['days_since_last_review'].astype(int)
```

```python
m = []
for i in range(len(df)):
    if pd.isna(df['reviews_per_month'][i]):
        m.append(np.NaN)
    else:
        m.append(df['number_of_reviews'][i]/df['reviews_per_month'][i])
df['month_listed'] = m
```

```python
df.head()
```

```python
df['days_since_last_review'].fillna(np.nanmedian(df['days_since_last_review']),inplace = True)
df['month_listed'].fillna(0,inplace = True)
#order
df['room_type'].replace(to_replace=['Entire home/apt','Private room', 'Shared room'], value=[3,2,1],inplace=True)
df = df.drop(columns = ['last_review','reviews_per_month','calculated_host_listings_count', 'availability_365'])
df = df[df['price'] > 10]
df = df[df['price'] < 1000]
df = df.reset_index()
corr = df.corr()

ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True,
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
)
```

```python
y = np.log(df.price)
features = df.drop(columns = ['price'])
features = pd.concat([features,pd.get_dummies(df['neighbourhoods'])], axis=1)
features = features.drop(columns = ['neighbourhoods', 'latitude', 'longitude'])
```

```python
count, bin_edges = np.histogram(y,bins = 50)
ax = y.plot(kind = 'hist', figsize = (8,5),bins = 20)

ax.set_title('Histogram of Airbnb Prices')
ax.set_ylabel('Number of Properties')
ax.set_xlabel('Price ($)')
```

```python
#check the relationship between room types and listed prices
num_properties_rt = df['room_type'].value_counts()
avg_price_rt = (df.groupby(['room_type'],as_index = False).mean())[['room_type','price']]
avg_price_rt = avg_price_rt.set_index(['room_type'])
ax = avg_price_rt.plot(kind = 'bar', figsize = (8,5))
ax.set_xticklabels(avg_price_rt.index, rotation = 45)
ax.set_title('Average Price Per Night ')
ax.set_ylabel('$')
ax.set_xlabel('Room Type')
```

```python
#Wordcloud, availability map
stopwords = set(STOPWORDS)
name_words = ' '
for string in df.name: 
      
    # typecaste each val to string 
    val = str(string) 
  
    # split the value 
    tokens = val.split() 
      
    # Converts each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
          
    for words in tokens: 
        name_words = name_words + words + ' '
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(name_words) 
```

```python
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```

## Modelling


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold

scaler = MinMaxScaler(feature_range=(0, 1))
features = scaler.fit_transform(features)

#X_train, X_test, y_train, y_test = train_test_split(features, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
scores = []
linear_model = LinearRegression()
cv = KFold(n_splits = 5, random_state = 0, shuffle=False)
for train_index, test_index in cv.split(features):
    X_train, X_test, y_train, y_test = features[train_index], features[test_index], y[train_index], y[test_index]
    linear_model.fit(X_train, y_train)
    linear_pred = linear_model.predict(X_test)
    linear_RMSE = np.sqrt(mean_squared_error(linear_pred,y_test))
    scores.append(linear_RMSE)
linear_score = np.mean(scores)
```

```python
linear_score
```

```python
from sklearn.linear_model import Lasso
lasso_scores = np.zeros(100)
lamda = list(np.linspace(0.00001,0.0001,100))
for i in range(len(lasso_scores)):
    scores = []
    lasso_model = Lasso(alpha = lamda[i])
    cv = KFold(n_splits = 5, random_state = 0, shuffle=False)
    for train_index, test_index in cv.split(features):
        X_train, X_test, y_train, y_test = features[train_index], features[test_index], y[train_index], y[test_index]
        lasso_model.fit(X_train, y_train)
        lasso_pred = lasso_model.predict(X_test)
        lasso_RMSE = np.sqrt(mean_squared_error(lasso_pred,y_test))
        scores.append(lasso_RMSE)
    lasso_score = np.mean(scores)
    lasso_scores[i] = lasso_score
    

```

```python
plt.plot(lamda, lasso_scores)
```

```python
lasso_model = Lasso(alpha = 0.1)
lasso_model.fit(X_train, y_train)
lasso_pred = lasso_model.predict(X_test)
lasso_RMSE = np.sqrt(mean_squared_error(lasso_pred,y_test))
```

```python
lasso_RMSE
```

```python
from sklearn.linear_model import Ridge

ridge_scores = np.zeros(100)
lamda = list(np.linspace(0.1,10,100))
for i in range(len(ridge_scores)):
    scores = []
    ridge_model = Ridge(alpha = lamda[i])
    cv = KFold(n_splits = 5, random_state = 0, shuffle=False)
    for train_index, test_index in cv.split(features):
        X_train, X_test, y_train, y_test = features[train_index], features[test_index], y[train_index], y[test_index]
        ridge_model.fit(X_train, y_train)
        ridge_pred = ridge_model.predict(X_test)
        ridge_RMSE = np.sqrt(mean_squared_error(ridge_pred,y_test))
        scores.append(ridge_RMSE)
    ridge_score = np.mean(scores)
    ridge_scores[i] = ridge_score
```

```python
np.min(ridge_scores)
```

```python
plt.plot(lamda,ridge_scores)
plt.xlabel('lambda')
plt.ylabel('RMSE')
```

```python
ridge_model = Ridge(alpha = 0.)
ridge_model.fit(X_train, y_train)
ridge_pred = ridge_model.predict(X_test)
ridge_RMSE = np.sqrt(mean_squared_error(ridge_pred,y_test))
```
